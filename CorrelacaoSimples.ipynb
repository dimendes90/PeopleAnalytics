```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Import NumPy library
print ("Ok Parça")
```

```python
# Importação dos dados
df = pd.read_excel("Correlacao_eNPS.xlsx")
```

```python
# Seleção das variáveis
x = df["enps"]
y = df["adesao"]
```

```python
# Cálculo da correlação
correlacao = df['enps'].corr(df['adesao'])
```

```python
# Plotar o gráfico de dispersão
plt.scatter(x, y)
# Adicionar linha de regressão linear
plt.plot(x, np.poly1d(np.polyfit(x, y, 1))(x), color='red')

# Adicionar título e rótulos aos eixos
plt.title("Correlação entre {} e {}".format(x.name, y.name))
plt.xlabel(x.name)
plt.ylabel(y.name)

# Exibir o coeficiente de correlação no gráfico
plt.annotate("Correlação: {:.2f}".format(correlacao), xy=(0.8, 0.8), xycoords='axes fraction')
# Mostrar o gráfico
plt.show()
```
